```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Set the output length
pd.options.display.max_rows = 500
```

```python
#This function can be used to find specific substring in a column. 
def search_in_column(data, look_in_column, output_columns, substring, show):
    data["temp"] = data[look_in_column].str.find(substring)
    print(data[output_columns][data["temp"] != -1][:show])
    return (data[output_columns][data["temp"] != -1][:show])
```

```python
columns = ['_id', 'date_updated', 'app.description', 'app.image_url', 'app.id', 'app.name', 'comments.count', 'payment.status', 'payment.id', 'payment.date_completed', 'payment.target.user.username', 'payment.target.user.last_name', 'payment.target.user.is_group', 'payment.target.user.is_active', 'payment.target.user.profile_picture_url', 'payment.target.user.is_blocked', 'payment.target.user.id', 'payment.target.user.date_joined', 'payment.target.user.display_name', 'payment.target.user.first_name', 'payment.target.type', 'payment.audience', 'payment.actor.username', 'payment.actor.last_name', 'payment.actor.is_group', 'payment.actor.is_active', 'payment.actor.profile_picture_url', 'payment.actor.is_blocked', 'payment.actor.id', 'payment.actor.date_joined', 'payment.actor.display_name', 'payment.actor.first_name', 'payment.action', 'payment.date_created', 'note', 'audience', 'likes.count', 'mentions.count', 'date_created', 'type', 'id']
chunk_size = 1000000
nrows = 4000000
df_chunk = pd.read_csv("venmo.csv", nrows = nrows, usecols = columns, chunksize=chunk_size)
print(columns)
print(len(columns))
```

```python
data_list = []
i = 0

for chunk in df_chunk:
    data_list.append(chunk)
    print(i)
    i = i + 1
```

## Find the people with the most transactions

```python
list_names_actor = []
list_names_target = []

for data in data_list:
    list_names_actor = list_names_actor + list(data["payment.actor.username"].value_counts().index[:50])
    list_names_target = list_names_target + list(data["payment.target.user.username"].value_counts().index[:50])
    
print(list_names_actor)
print("\n")
print(list_names_target)
```

```python
#Create names list which is the combination of the list_names_actor and list_names_target without duplicates
set_1 = set(list_names_actor)
set_2 = set(list_names_target)

names_list = list_names_actor + list(set_2 - set_1)
print(names_list)
print(len(names_list))
```

```python
#This cell drops all the columns which are not included in the list
for i in range(len(data_list)):
    data_list[i]["names_to_keep"] = data_list[i]["payment.actor.username"].isin(names_list) | data_list[i]["payment.target.user.username"].isin(names_list)
    data_list[i] = data_list[i][data_list[i]["names_to_keep"] == True]
```

```python
#Concatenate the data
final_data = pd.concat(data_list, axis = 0)
```
